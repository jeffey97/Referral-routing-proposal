```python
import numpy as np
import matplotlib.pyplot as plt

# Example hospital loads
hospital_loads = np.array([100, 120, 80, 110])  # Current patient loads at hospitals

# Calculate the average load
average_load = np.mean(hospital_loads)

# Calculate load balancing metric (standard deviation)
load_balancing_metric = np.sqrt(np.mean((hospital_loads - average_load)**2))  # Standard deviation of loads

# Calculate relative load balancing (coefficient of variation)
coefficient_of_variation = load_balancing_metric / average_load * 100  # Expressed as a percentage

# Find the hospital with the maximum deviation from average load
deviation_from_average = hospital_loads - average_load
max_deviation_hospital = np.argmax(np.abs(deviation_from_average)) + 1  # Add 1 for hospital index (1-based)

# Output results
print("\n--- Load Balancing Analysis ---")
print(f"Hospital Loads: {hospital_loads}")
print(f"Average Load: {average_load:.2f}")
print(f"Load Balancing Metric (Standard Deviation): {load_balancing_metric:.2f}")
print(f"Coefficient of Variation: {coefficient_of_variation:.2f}%")
print(f"Hospital with Maximum Deviation from Average Load: Hospital {max_deviation_hospital}")

# Visualization of load distribution and balance
plt.figure(figsize=(10, 6))
hospitals = range(1, len(hospital_loads) + 1)

# Bar plot for hospital loads
plt.bar(hospitals, hospital_loads, alpha=0.7, label="Hospital Load", color="skyblue")
plt.axhline(y=average_load, color="red", linestyle="--", label=f"Average Load ({average_load:.2f})")

# Highlight the hospital with the maximum deviation
plt.bar(max_deviation_hospital, hospital_loads[max_deviation_hospital - 1], color="orange", alpha=0.9, label="Max Deviation")

# Labels and Title
plt.xlabel("Hospital")
plt.ylabel("Patient Load")
plt.title("Hospital Load Balancing Analysis")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.xticks(hospitals)
plt.show()
```
